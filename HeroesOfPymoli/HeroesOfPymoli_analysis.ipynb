```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# Create player_count variable to filter the unique number of players
player_count_df = purchase_data.drop_duplicates(subset=['SN'])
player_count = player_count_df.count()
# Create a summary data frame to show total number of players
total_player_df = pd.DataFrame({'Total Players': [player_count['SN']]})
total_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Create variables separately to store the calculations
number_of_unique_items = purchase_data.drop_duplicates(subset=['Item Name']).count()
average_price = purchase_data['Price'].mean()
number_of_purchases = purchase_data['Item Name'].count()
total_revenue = purchase_data['Price'].sum()

# Create summary data frame to show the results
purchasing_analysis = pd.DataFrame({
    'Number of Unique Items': [number_of_unique_items['Item Name']],
    'Average Price': '${:,.2f}'.format(average_price),
    'Number of Purchases': [number_of_purchases],
    'Total Revenue': '${:,.2f}'.format(total_revenue)
})

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Use the 'player_count_df' data frame that was previously created to calculate total number of players (this data frame only includes the non-duplicated players) to arrange the data by Gender
gender_df = player_count_df.set_index('Gender')
player_count_df

# Create variables to store gender demographics calculations
male_count = gender_df.loc['Male','SN'].count()
female_count = gender_df.loc['Female','SN'].count()
other_count = gender_df.loc['Other / Non-Disclosed','SN'].count()

total_players = player_count['SN']     # This variable was used in the Total Number of Players calculations
male_pct = male_count / total_players
female_pct = female_count / total_players
other_pct = other_count / total_players

# Create summary data frame
data_gender = {
    'Total Count': [male_count, female_count, other_count],
    'Percentage of Players': [male_pct, female_pct, other_pct]
}
gender_demo = pd.DataFrame(data_gender, index=['Male',
                                        'Female',
                                        'Other / Non-Disclosed'])

# Format the percentages
gender_demo['Percentage of Players'] = gender_demo['Percentage of Players'].map('{:.2%}'.format)
gender_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create age groups using the original dataset
age_groups = pd.cut(
    x=purchase_data['Age'],
    bins=age_to_cut,
    labels=age_group_name,
    include_lowest=True,
)

purchase_count = purchase_data.groupby(age_groups)['Item Name'].count()
purchase_total = purchase_data.groupby(age_groups)['Price'].sum()

purchase_avg = purchase_total / purchase_count
purchase_avg_person = purchase_total / age_count

    
# Create summary data frame
data_purchase = {
    'Purchase Count': purchase_count,
    'Average Purchase Price': purchase_avg,
    'Total Purchase Value': purchase_total,
    'Avg Total Purchase per Person': purchase_avg_person
}
purchase_demo = pd.DataFrame(data_purchase, index=age_group_name)

# Format the floats
purchase_demo['Average Purchase Price'] = purchase_demo['Average Purchase Price'].map('${:,.2f}'.format)
purchase_demo['Total Purchase Value'] = purchase_demo['Total Purchase Value'].map('${:,.2f}'.format)
purchase_demo['Avg Total Purchase per Person'] = purchase_demo['Avg Total Purchase per Person'].map('${:,.2f}'.format)
purchase_demo
    
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create spender_df data frame to make 'SN' as index
spender_df = purchase_data.set_index('SN')

# Add the analysis columns to the data frame
spender_count = spender_df.groupby(['SN'])['Item Name'].count()
spender_total = spender_df.groupby(['SN'])['Price'].sum()
spender_avg = spender_total / spender_count

spender_df['Purchase Count'] = spender_count
spender_df['Average Purchase Price'] = spender_avg
spender_df['Total Purchase Value'] = spender_total

# Create summary data frame
spender_df = spender_df.drop(columns=['Purchase ID', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'], axis=1)
spender_demo = spender_df.sort_values('Total Purchase Value', ascending=False).drop_duplicates(subset=['Total Purchase Value'])

# Format the floats
spender_demo['Average Purchase Price'] = spender_demo['Average Purchase Price'].map('${:,.2f}'.format)
spender_demo['Total Purchase Value'] = spender_demo['Total Purchase Value'].map('${:,.2f}'.format)
spender_demo.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create spender_df data frame to make 'Item ID' as index
popular_df = purchase_data.set_index('Item ID')

# Add the analysis columns to the data frame
popular_count = popular_df.groupby(['Item ID'])['Purchase ID'].count()
popular_total = popular_df.groupby(['Item ID'])['Price'].sum()
popular_price = popular_total / popular_count

popular_df['Purchase Count'] = popular_count
popular_df['Item Price'] = popular_price
popular_df['Total Purchase Value'] = popular_total

# Create summary data frame
popular_df = popular_df.drop(columns=['Purchase ID', 'Age', 'Gender', 'SN', 'Price'], axis=1)
popular_items = popular_df.sort_values('Purchase Count', ascending=False).drop_duplicates(subset=['Total Purchase Value'])

# Format the floats
popular_items['Item Price'] = popular_items['Item Price'].map('${:,.2f}'.format)
popular_items['Total Purchase Value'] = popular_items['Total Purchase Value'].map('${:,.2f}'.format)
popular_items.head(10)

```
